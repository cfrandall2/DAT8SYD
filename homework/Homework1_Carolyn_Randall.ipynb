```python
# Import the libraries required
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt
# this allows plots to appear directly in the notebook
%matplotlib inline
```

```python
# Read in data
#movie_data = pd.read_csv("u.user", index_col=0)
movie_data = pd.read_csv("u.user", delimiter='|', names=('row', 'age', 'gender', 'occupation', 'numberid'))
#dataset = pd.read_csv(raw_data, delimiter=",", names=('sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species'))

# display to confirm data read ok
movie_data.head()
```

```python
#describe the data
movie_data.describe()
```

```python
byoccupation = movie_data.groupby('occupation')
byoccupation.describe()
```

```python
#count the number of occurances for each occupation in the file
byoccupation.count()
```

```python
# for each occupation, calculate the mean age
#byspecies['petal_length'].agg([len, np.mean, np.std])
byoccupation['age'].aggregate(np.mean)
# 
```

```python
# for each occupation, calculate the minimum age
byoccupation['age'].aggregate(np.min)
```

```python
# for each occupation, calculate the maximum age
byoccupation['age'].aggregate(np.max)
```

```python
# for each combination of occupation and gender, calculate the mean age
byOccAndGender = movie_data.groupby(['occupation', 'gender'])

#byoccupation = movie_data.groupby('occupation')
byOccAndGender.describe()
```

```python
# for each combination of occupation and gender, calculate the mean age
byOccAndGender.aggregate(np.mean)
```
